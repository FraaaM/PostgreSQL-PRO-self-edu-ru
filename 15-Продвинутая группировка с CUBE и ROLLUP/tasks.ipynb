
## Домашнее Задание "Продвинутые группировки"

### Задача 1:

Вы — аналитик в компании Northwind. Руководство просит вас подготовить сводный отчёт по эффективности работы транспортных компаний (`Shippers`).

**Требования к отчёту:**
Отчёт должен быть **единым SQL-запросом** и содержать следующие показатели, рассчитанные для **каждого года**:

1.  Общая стоимость фрахта (`freight`) для каждой транспортной компании (`shipper`).
2.  Общая стоимость фрахта по всем компаниям (подитог для каждого года).
3.  Общий итог по фрахту за всё время по каждой транспортной компании (без учёта года).
4.  Глобальный итог по фрахту за весь период.

**Подсказка:** Подумайте, какая из конструкций (`GROUPING SETS`, `ROLLUP`, `CUBE`) наиболее точно соответствует этим требованиям. Вам понадобятся таблицы `orders` и `shippers`.

`ROLLUP` идеально подходит для данных с **иерархией**. Классический пример — это временные данные: `Год -> Квартал -> Месяц`

`CUBE` используется для **многомерного анализа**, когда нет строгой иерархии, и вам нужны итоги по всем возможным комбинациям полей. Представим, что вы хотите проанализировать, какие товары и в какие страны поставляются.

### Задача 2:

Напишите **один SQL-запрос**, который посчитает количество нанятых сотрудников (`employees`) и выведет:
1.  Количество сотрудников, нанятых в каждый конкретный год и квартал.
2.  Подитог по каждому году.
3.  Общий итог по всем нанятым сотрудникам за всё время.

**Таблицы для использования:** `employees`
**Поля для использования:** `hire_date`

### Задача 3:

Напишите **один SQL-запрос**, который посчитает общую стоимость заказанных товаров (`unit_price * quantity`) и выведет:
1.  Сумму по каждой **категории товаров** (`category_name`), отправленную в каждую **страну** (`ship_country`).
2.  Подитог по каждой категории товаров (по всем странам).
3.  Подитог по каждой стране (по всем категориям).
4.  Общий итог по всем заказам.

**Таблицы для использования:** `orders`, `order_details`, `products`, `categories`.
**Поля для использования:** `ship_country` из `orders`, `category_name` из `categories`, `unit_price` и `quantity` из `order_details`.
