### ДЗ "Массивы"

#### Условие

Вы работаете в отделе логистики компании Northwind. Для оптимизации поставок вам нужно решить несколько задач с помощью массивов.

**Задача 1: Регионы обслуживания**
Поставщики (`suppliers`) теперь обслуживают не один город, а несколько. Вам нужно добавить в таблицу `suppliers` информацию о городах, в которые они осуществляют доставку.

1.  Добавьте в таблицу `suppliers` новую колонку `shipping_cities` с типом `TEXT[]`.
2.  Для поставщика с `supplier_id = 1` (`Exotic Liquids`) добавьте три города доставки: 'London', 'Berlin', 'Madrid'.
3.  Для поставщика с `supplier_id = 2` (`New Orleans Cajun Delights`) добавьте города 'New Orleans' и 'Chicago'.
4.  Для поставщика с `supplier_id = 3` (`Grandma Kelly's Homestead`) добавьте города 'Ann Arbor', 'Chicago', 'Detroit'.

**Задача 2: Анализ поставок**
Теперь, используя новые данные, напишите несколько запросов:

1.  Найдите всех поставщиков, которые доставляют товары в 'Chicago'.
2.  Найдите всех поставщиков, которые доставляют и в 'London', и в 'Berlin' (должны присутствовать оба города).
3.  Найдите всех поставщиков, чей список городов доставки содержит 'Detroit', но не содержит 'New Orleans'.
4.  К списку городов поставщика `supplier_id = 1` добавьте новый город 'Paris', не удаляя старые. Напишите `UPDATE` запрос.

### ДЗ "Массивы и Циклы"

**Задача 1: Функция для пакетного обновления цен**

Отдел маркетинга предоставил вам список ID товаров (`product_id`), на которые нужно срочно установить скидку для проведения рекламной кампании. Ваша задача — написать универсальную функцию, которая принимает этот список (в виде массива) и применяет скидку.

Напишите функцию `ApplyDiscountByProductList(product_ids_to_update INT[], discount_percentage NUMERIC)`.

**Логика функции:**

1.  Функция должна принимать два аргумента:
      * `product_ids_to_update`: **массив** целочисленных ID товаров.
      * `discount_percentage`: процент скидки (например, 12.5 для 12.5%).
2.  Используя цикл `FOREACH`, переберите каждый `product_id` из входного массива `product_ids_to_update`.
3.  Внутри цикла для каждого ID товара:
      * Получите его текущую цену (`unit_price`) из таблицы `products`.
      * Рассчитайте новую цену, уменьшенную на `discount_percentage`.
      * Обновите цену для этого товара в таблице `products`.
4.  Функция должна возвращать таблицу с колонками `product_id`, `product_name`, `old_price`, `new_price`, чтобы показать, какие именно товары были обновлены и как изменилась их цена.
5.  Добавьте проверку: если в переданном массиве окажется ID, которого нет в таблице `products`, функция должна проигнорировать его и продолжить работу с остальными, а не прерываться с ошибкой.

#### ДОП. ЗАДАЧА (без массивов)

Отдел продаж хочет внедрить систему скидок на товары, которые плохо продаются. Вам поручено написать функцию на `PL/pgSQL`, которая автоматизирует этот процесс.

**Задача 2: Функция для уценки товаров**

Напишите функцию `ApplyDiscountForSlowMovingProducts(max_quantity_sold INT, discount_percentage NUMERIC)`.

**Логика функции:**

1.  Функция должна принимать два аргумента:
      * `max_quantity_sold`: максимальное количество проданных единиц. Если товар продавался в количестве, не превышающем это значение, он считается "плохо продаваемым".
      * `discount_percentage`: процент скидки (например, 10.0 для 10%).
2.  Функция должна найти все товары (`products`), суммарное количество продаж которых (из `order_details`) **меньше или равно** `max_quantity_sold`.
3.  Используя цикл (`FOR` или `FOREACH`), нужно перебрать найденные ID товаров.
4.  Внутри цикла для каждого такого товара нужно обновить его цену (`unit_price`), уменьшив её на указанный процент. Убедитесь, что скидка не превышает 50% от первоначальной цены (добавьте проверку).
5.  Функция должна возвращать таблицу с колонками `product_id`, `product_name`, `old_price`, `new_price` для всех товаров, чья цена была обновлена.
